```python
import pandas as pd
import numpy as np
from urllib.request import urlopen, Request

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36'}
reg_url = "https://diagnos.co.id/2020/03/21/data-realtime-kasus-covid-19-di-seluruh-dunia/"
req = Request(url=reg_url, headers=headers) 
html = urlopen(req).read() 

df_list = pd.read_html(html)
df_list
```

```python
len(df_list)
```

```python
df0 = df_list[0]
df0
```

```python
df0.dtypes
```

```python
Positif_new = (round(df0.loc[0:8, ['Positif']] * 1000)).append(df0.loc[9:, ['Positif']])
Positif_new
```

```python
Sembuh_new = (round(df0.loc[0:3,['Sembuh']] * 1000)).append(df0.loc[4:,['Sembuh']])
Sembuh_new
```

```python
df1 = df0.copy()
df1.head()
```

```python
df1 = df0.copy()
df1['Positif']= Positif_new
df1['Sembuh'] = Sembuh_new
df1
```

```python
cols = ['Positif','Sembuh']
df1[cols] = df1[cols].applymap(np.int64)
df1
```

```python
import datetime as dt
date = dt.date.today().strftime("%d/%m/%Y")

df1['Last Update'] = date
df1
```
